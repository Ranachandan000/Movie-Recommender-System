```python
import ast 
def extract(object):
    list = []
    for i in ast.literal_eval(object) :
        list.append(i['name'])
    return list    
```

```python
df['genres'] = df['genres'].apply(extract)
```

```python
df.head(1)
```

```python
#doing the same with 'keywords' column
```

```python
df['keywords'] = df['keywords'].apply(extract)
```

```python
# for cast
df['cast'][0]  # will take only lead characters upto 5
```

```python
def extract2(text):
    List = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5:
            List.append(i['name'])
        counter+=1
    return List
```

```python
df['cast'] = df['cast'].apply(extract2) 
```

```python
df.head(1)
```

```python
df['crew'].unique()
```

```python
def fetch_director(text):
    list = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            list.append(i['name'])
    return list
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
df.head()
```

```python
type(df['overview'][0])       # gotta turn into a list
```

```python
def Convert(string):
    li = list(string.split(" "))
    return li
```

```python
df['overview'] = df['overview'].apply(Convert)
```

```python
df.head(1)
```

```python
#to remove the confusion among two 'Sam' , we'll joint a cast's whole name.
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1


df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
```

```python
df.head(1)
```

```python
df['tags'] = df['overview'] + df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df.head(1)
```

```python
df.drop(['overview','genres','keywords','cast','crew'] , axis = 1 , inplace = True)
```

```python
df.head(1)
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df.head(1)
```

```python
df['tags'][0]
```

```python
df['tags'] = df['tags'].apply(lambda x : x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
